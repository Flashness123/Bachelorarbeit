```python
from pdf2image import convert_from_path
from PIL import Image
import os

pdfs_dir = r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\Alle_Daten'
pngs_dir = r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\Alle_Daten\PNG_Output'

# Ensure the PNGs directory exists
os.makedirs(pngs_dir, exist_ok=True)

Image.MAX_IMAGE_PIXELS = None

for pdf_file in os.listdir(pdfs_dir):
    if pdf_file.endswith('.pdf'):
        pdf_path = os.path.join(pdfs_dir, pdf_file)
        try:
            images = convert_from_path(pdf_path)
            for i, image in enumerate(images):
                # Define the path for saving PNGs, making sure to save them in the new directory
                png_path = os.path.join(pngs_dir, f'{pdf_file[:-4]}_page_{i+1}.png')
                image.save(png_path, 'PNG')
        except Exception as e:
            print(f"Error converting {pdf_file}: {e}")
```

## Deleteiong all pages higher 1 and renaming the ones with 1

```python
import os
import glob

# Set the directory where your files are located
directory = r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\PNG_Output'

# Pattern to match files ending with _page_1.png
page_1_files = glob.glob(os.path.join(directory, '*_page_1.png'))

# Rename _page_1 files
for filepath in page_1_files:
    # Split the path and the extension
    base_name, extension = os.path.splitext(filepath)
    # Remove the _page_1 part and keep the extension
    new_name = base_name.rsplit('_page_1', 1)[0] + extension
    os.rename(filepath, new_name)
    print(f'Renamed: {filepath} to {new_name}')

# Pattern to match files ending with _page_2 or higher and with .png extension
page_2_and_higher_files = glob.glob(os.path.join(directory, '*_page_[2-9].png')) + glob.glob(os.path.join(directory, '*_page_[1-9][0-9].png'))

# Delete _page_2 and higher files
for filepath in page_2_and_higher_files:
    os.remove(filepath)
    print(f'Deleted: {filepath}')
```

## merging into 1 subdirectory

```python
import shutil
import os

def copy_pngs_with_jsons(json_dir, png_dir, target_dir):
    """
    Copies PNG images from png_dir to target_dir if there is a corresponding JSON file in json_dir.
    
    Args:
    - json_dir: Directory containing JSON files.
    - png_dir: Directory containing PNG images.
    - target_dir: Directory to copy the PNG images that have corresponding JSON files.
    """
    
    # Ensure the target directory exists
    os.makedirs(target_dir, exist_ok=True)
    
    # List all PNG files in the PNG directory
    for png_file in os.listdir(png_dir):
        if png_file.lower().endswith('.png'):
            # Construct the expected JSON filename based on the PNG filename
            base_name = os.path.splitext(png_file)[0]
            json_filename = base_name + '.pdf.labels.json'
            
            # Construct the full paths for the JSON file and the source PNG file
            json_path = os.path.join(json_dir, json_filename)
            png_path = os.path.join(png_dir, png_file)
            
            # Check if the JSON file exists
            if os.path.exists(json_path):
                # Copy the PNG file to the target directory
                shutil.copy(png_path, target_dir)
                print(f"Copied {png_file} to {target_dir}")

# Example usage
json_dir =   r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\png_with_label'
png_dir =    r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\PNG_Output'
target_dir = r'C:\Users\lukask\Coding\Interface Projects\Table_Recognition\Complete_Data\png_with_label'  # Replace with the actual target directory path

copy_pngs_with_jsons(json_dir, png_dir, target_dir)
```
